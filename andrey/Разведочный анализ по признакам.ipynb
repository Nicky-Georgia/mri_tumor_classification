```python

from PIL import Image
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 
glioma_folder = "C:\\Users\\vorob\\Downloads\\Dataset_full\\Training\\glioma"
meningioma_folder = "C:\\Users\\vorob\\Downloads\\Dataset_full\\Training\\meningioma"
pituitary_folder = "C:\\Users\\vorob\\Downloads\\Dataset_full\\Training\\pituitary"
notumor_folder = "C:\\Users\\vorob\\Downloads\\Dataset_full\\Training\\notumor"

# 
glioma_brightness_values = []
meningioma_brightness_values = []
pituitary_brightness_values = []
notumor_brightness_values = []

glioma_texture_values = []
meningioma_texture_values = []
pituitary_texture_values = []
notumor_texture_values = []

glioma_size_values = []
meningioma_size_values = []
pituitary_size_values = []
notumor_size_values = []

glioma_density_values = []
meningioma_density_values = []
pituitary_density_values = []
notumor_density_values = []
```

```python

#  
# Glioma
for image_name in os.listdir(glioma_folder):
 
    image_path = os.path.join(glioma_folder, image_name)
    img = Image.open(image_path)

    img_array = np.array(img)

    brightness = np.mean(img_array)
    texture = np.std(img_array)
    size = img_array.shape[0] * img_array.shape[1]
    density = np.count_nonzero(img_array) / size

    glioma_brightness_values.append(brightness)
    glioma_texture_values.append(texture)
    glioma_size_values.append(size)
    glioma_density_values.append(density)

# Meningioma
for image_name in os.listdir(meningioma_folder):

    image_path = os.path.join(meningioma_folder, image_name)
    img = Image.open(image_path)

    img_array = np.array(img)

    meningioma_brightness = np.mean(img_array)
    meningioma_texture = np.std(img_array)
    meningioma_size = img_array.shape[0] * img_array.shape[1]
    meningioma_density = np.count_nonzero(img_array) / size

    
    meningioma_brightness_values.append(meningioma_brightness)
    meningioma_texture_values.append(meningioma_texture)
    meningioma_size_values.append(meningioma_size)
    meningioma_density_values.append(meningioma_density)
    
# Pituitary   
for image_name in os.listdir(pituitary_folder):
 
    image_path = os.path.join(pituitary_folder, image_name)
    img = Image.open(image_path)

    img_array = np.array(img)

    pituitary_brightness = np.mean(img_array)
    pituitary_texture = np.std(img_array)
    pituitary_size = img_array.shape[0] * img_array.shape[1]
    pituitary_density = np.count_nonzero(img_array) / size

    pituitary_brightness_values.append(pituitary_brightness)
    pituitary_texture_values.append(pituitary_texture)
    pituitary_size_values.append(pituitary_size)
    pituitary_density_values.append(pituitary_density)
    
# Notumor   
for image_name in os.listdir(notumor_folder):
    
    image_path = os.path.join(notumor_folder, image_name)
    img = Image.open(image_path)

    img_array = np.array(img)

    notumor_brightness = np.mean(img_array)
    notumor_texture = np.std(img_array)
    notumor_size = img_array.shape[0] * img_array.shape[1]
    notumor_density = np.count_nonzero(img_array) / size

    notumor_brightness_values.append(notumor_brightness)
    notumor_texture_values.append(notumor_texture)
    notumor_size_values.append(notumor_size)
    notumor_density_values.append(notumor_density)
```

```python
# DataFrame
glioma_pixel_values = pd.DataFrame({
    'brightness_glioma': glioma_brightness_values,
    'texture_glioma': glioma_texture_values,
    'size_glioma': glioma_size_values,
    'density_glioma': glioma_density_values
})

meningioma_pixel_values = pd.DataFrame({
    'brightness_meningioma': meningioma_brightness_values,
    'texture_meningioma': meningioma_texture_values,
    'size_meningioma': meningioma_size_values,
    'density_meningioma': meningioma_density_values
})

pituitary_pixel_values = pd.DataFrame({
    'brightness_pituitary': pituitary_brightness_values,
    'texture_pituitary': pituitary_texture_values,
    'size_pituitary': pituitary_size_values,
    'density_pituitary': pituitary_density_values
})

notumor_pixel_values = pd.DataFrame({
    'brightness_notumor': notumor_brightness_values,
    'texture_notumor': notumor_texture_values,
    'size_notumor': notumor_size_values,
    'density_notumor': notumor_density_values
})
```

```python
#  
# Brightness
#  'brightness_glioma'
plt.hist(glioma_pixel_values['brightness_glioma'], bins=50, alpha=0.7, color='r')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('brightness value distribution (class: glioma)')
plt.show()

#   'brightness_meningioma'
plt.hist(meningioma_pixel_values['brightness_meningioma'], bins=50, alpha=0.7, color='g')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('Brightness value distribution (class: meningioma)')
plt.show()

#   'brightness_pitutiary'
plt.hist(pituitary_pixel_values['brightness_pituitary'], bins=50, alpha=0.7, color='b')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('Brightness value distribution (class: pituitary)')
plt.show()

#  'brightness_notumor'
plt.hist(notumor_pixel_values['brightness_notumor'], bins=50, alpha=0.7, color='black')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('Brightness value distribution (class: notumor)')
plt.show()
```
