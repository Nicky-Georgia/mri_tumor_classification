```python
# Import modules

from PIL import Image
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Folders

glioma_folder = '../data/Testing/glioma'
meningioma_folder = '../data/Testing/meningioma'
pituitary_folder = '../data/Testing/pituitary'
notumor_folder = '../data/Testing/notumor'

# Init lists

glioma_brightness_values = []
meningioma_brightness_values = []
pituitary_brightness_values = []
notumor_brightness_values = []

glioma_texture_values = []
meningioma_texture_values = []
pituitary_texture_values = []
notumor_texture_values = []

glioma_size_values = []
meningioma_size_values = []
pituitary_size_values = []
notumor_size_values = []

glioma_density_values = []
meningioma_density_values = []
pituitary_density_values = []
notumor_density_values = []
```

```python
# Process folder function

def process_folder(folder_path, brightness_values, texture_values,
                   size_values, density_values):
    for image_name in os.listdir(folder_path):
        image_path = os.path.join(folder_path, image_name)
        img = Image.open(image_path)
        img_array = np.array(img)

        brightness = np.mean(img_array)
        texture = np.std(img_array)
        size = img_array.shape[0] * img_array.shape[1]
        density = np.count_nonzero(img_array) / size

        brightness_values.append(brightness)
        texture_values.append(texture)
        size_values.append(size)
        density_values.append(density)
```

```python
# Process folders

process_folder(glioma_folder, glioma_brightness_values,
               glioma_texture_values, glioma_size_values,
               glioma_density_values)

process_folder(meningioma_folder, meningioma_brightness_values,
               meningioma_texture_values, meningioma_size_values,
               meningioma_density_values)

process_folder(pituitary_folder, pituitary_brightness_values,
               pituitary_texture_values, pituitary_size_values,
               pituitary_density_values)

process_folder(notumor_folder, notumor_brightness_values,
               notumor_texture_values, notumor_size_values,
               notumor_density_values)
```

```python
# Create DataFrame

glioma_pixel_values = pd.DataFrame({
    'brightness_glioma': glioma_brightness_values,
    'texture_glioma': glioma_texture_values,
    'size_glioma': glioma_size_values,
    'density_glioma': glioma_density_values
})

meningioma_pixel_values = pd.DataFrame({
    'brightness_meningioma': meningioma_brightness_values,
    'texture_meningioma': meningioma_texture_values,
    'size_meningioma': meningioma_size_values,
    'density_meningioma': meningioma_density_values
})

pituitary_pixel_values = pd.DataFrame({
    'brightness_pituitary': pituitary_brightness_values,
    'texture_pituitary': pituitary_texture_values,
    'size_pituitary': pituitary_size_values,
    'density_pituitary': pituitary_density_values
})

notumor_pixel_values = pd.DataFrame({
    'brightness_notumor': notumor_brightness_values,
    'texture_notumor': notumor_texture_values,
    'size_notumor': notumor_size_values,
    'density_notumor': notumor_density_values
})
```

```python
# Create Hist


# Brightness

# Hist 'brightness_glioma'
plt.hist(glioma_pixel_values['brightness_glioma'],
         bins=50, alpha=0.7, color='r')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('brightness value distribution (class: glioma)')
plt.show()

# Hist 'brightness_meningioma'
plt.hist(meningioma_pixel_values['brightness_meningioma'],
         bins=50, alpha=0.7, color='g')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('Brightness value distribution (class: meningioma)')
plt.show()

# Hist 'brightness_pitutiary'
plt.hist(pituitary_pixel_values['brightness_pituitary'],
         bins=50, alpha=0.7, color='b')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('Brightness value distribution (class: pituitary)')
plt.show()

# Hist 'brightness_notumor'
plt.hist(notumor_pixel_values['brightness_notumor'],
         bins=50, alpha=0.7, color='black')
plt.xlabel('Brightness')
plt.ylabel('Images number')
plt.title('Brightness value distribution (class: notumor)')
plt.show()
```

```python
# Параметр "яркость" различается среди различных классов (особенно сильно отличается класс notumor)
# В принципе прослеживается закономерность (яркость увеличивается от класса к классу)
# Хоть различия не особо большие

#Вероятно, закономерность будет более очевидной, если анализ проводить с сегментацией изображений
#Такой анализ заметно сложнее, но, вероятно, в дальнейшем он будет произведен
```
