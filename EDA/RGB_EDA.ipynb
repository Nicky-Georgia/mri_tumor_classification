# Выводы по результатам нахождения среднего и дисперсии по значениям пикселей по каждому из RGB фильтров:
Средние значения для классов опухолей заметно отличаются от класса "без опухоли", что свидетельствует о потенциальной дифференциации интенсивности цвета в разных классах. Дисперсии по классам опухолей несколько схожи, что говорит о схожих закономерностях распределения внутри каждого типа опухолей, но заметно отличаются от класса "без опухоли". Однако для того, чтобы сделать конкретные выводы о различиях или закономерностях между этими классами опухолей, потребуется более глубокий анализ и, возможно, другие характеристики, помимо цветовых каналов RGB. Данные наблюдения свидетельствуют о наличии различий в цветовых характеристиках изображений различных классов опухолей головного мозга, что потенциально может помочь в классификации или идентификации изображений.

```python
# Import modules

import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Functions for EDA

def calculate_mean_variance(images_folder):

    image_files = [f for f in os.listdir(images_folder)
                   if os.path.isfile(os.path.join(images_folder, f))]

    mean_values = {'R': [], 'G': [], 'B': []}
    variance_values = {'R': [], 'G': [], 'B': []}

    for img_file in image_files:
        img = cv2.imread(os.path.join(images_folder, img_file))
        b, g, r = cv2.split(img)

        for channel, channel_name in zip((r, g, b), ('R', 'G', 'B')):
            mean = np.mean(channel)
            variance = np.var(channel)
            mean_values[channel_name].append(mean)
            variance_values[channel_name].append(variance)

    overall_mean = {
        channel: np.mean(values) for channel, values in mean_values.items()
        }
    overall_variance = {
        channel: np.mean(values) for channel, values in variance_values.items()
        }

    return overall_mean, overall_variance


def plt_build(mean_values, veriance_values):

    plt.figure(figsize=(8, 4))
    plt.bar(mean_values.keys(), mean_values.values(),
            color=['red', 'green', 'blue'])
    plt.title('Mean Values for RGB Channels')
    plt.xlabel('Channels')
    plt.ylabel('Mean')
    plt.show()

    plt.figure(figsize=(8, 4))
    plt.bar(variance_values.keys(), variance_values.values(),
            color=['red', 'green', 'blue'])
    plt.title('Variance Values for RGB Channels')
    plt.xlabel('Channels')
    plt.ylabel('Variance')
    plt.show()
```

```python
# Draw plots and calculate mean and variance


# Pituitary

print('Pituitary')
folder_path = '../data/Training/pituitary'
means, variances = calculate_mean_variance(folder_path)
print("Mean values for each channel (R, G, B):", means)
print("Variance values for each channel (R, G, B):", variances)

mean_values = means
variance_values = variances
plt_build(mean_values, variance_values)


# Notumor

print('Notumor')
folder_path = '../data/Training/notumor'
means, variances = calculate_mean_variance(folder_path)
print("Mean values for each channel (R, G, B):", means)
print("Variance values for each channel (R, G, B):", variances)

mean_values = means
variance_values = variances
plt_build(mean_values, variance_values)


# Meningioma

print('Meningioma')
folder_path = '../data/Training/meningioma'
means, variances = calculate_mean_variance(folder_path)
print("Mean values for each channel (R, G, B):", means)
print("Variance values for each channel (R, G, B):", variances)

mean_values = means
variance_values = variances
plt_build(mean_values, variance_values)


# Glioma

print('Glioma')
folder_path = '../data/Training/glioma'
means, variances = calculate_mean_variance(folder_path)
print("Mean values for each channel (R, G, B):", means)
print("Variance values for each channel (R, G, B):", variances)

mean_values = means
variance_values = variances
plt_build(mean_values, variance_values)
```
